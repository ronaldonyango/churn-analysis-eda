### **Data-driven journey**

Our exploration begins with a dataset of 142,902 users. This rich information contains details on Unique ID, Date Joined, Gender, Age and Last active date. Before diving into analysis, we carefully cleaned and prepared the data, ensuring its accuracy and consistency.

### **Unveiling user demographics**

Who are our users? Examining demographic data, we discovered a surprising gender imbalance: the male population is 43% larger than the female population. This finding highlights the potential need for targeted marketing strategies.

### **Age and its influence**

Age plays a significant role in user behavior. The 20-24 age group dominates the user base, representing over half of all users. However, this seemingly dominant group also exhibits the highest churn rate. Understanding their needs and preferences becomes crucial for retaining them.

### **Time unveils hidden patterns**

Analyzing user join dates revealed intriguing temporal patterns. While December tends to see higher churn rates, users joining in August and October experience lower churn. This suggests tailoring onboarding experiences based on joining month could significantly impact retention.

### **A gender gap in churn**

One notable finding exposed a significant difference in churn rates between genders. Female users churn at a considerably higher rate compared to males and other genders. Investigating the reasons behind this disparity could unlock valuable insights and inform specific retention strategies.

### **The Overall Picture**

Combining all factors, the analysis revealed an overall churn rate of approximately 94%. While this number may seem daunting, it also presents a vast opportunity to improve user retention and engagement.

## **EXPLORATORY DATA ANALYSIS**

***

### **Data Preparation**

We will use Pandas to load the dataset into a dataframe, explore the different columns and ranges of values, handle missing values and incorrect datatypes, and generally prepare our data for analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns 
import matplotlib.ticker as mtick
```

We will load the dataset

```python
df = pd.read_excel("./data/dataset.xlsx")
```

We will view the basic information about our data using df.info() and df.describe()

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

***

**Data Cleaning**

```python
df = df.dropna(subset =['joined_date'])
```

Check for missing values in age and gender columns

```python
missing_percentages = round(df[['gender', 'age']].isna().sum() * 100 / len(df), 2)
missing_percentages
```

Fill missing values in 'gender' with 'Not provided' and in 'age' with the mean

```python
df.fillna({'gender': 'Not provided', 'age': df['age'].mean()}, inplace=True)
```

```python
df['age'] = df['age'].astype(int)
```

```python
gender_mapping = {'male': 'Male', 'female': 'Female', 'Other': 'Not provided'}
df['gender'] = df['gender'].replace(gender_mapping)
```

Convert the joined date column data type to datetime format

```python
df['joined_date'] = pd.to_datetime(df['joined_date'], errors='coerce')
```

Create 'churn' column indicating churned versus active users

```python
df['churn'] = np.where(df['last active date'] == "> 90 days", "Churned", "Active")
```

Create 'churn_rate' column with 1 for churned and 0 for active

```python
df['churn_rate'] = np.where(df['churn'] == "Churned", 1, 0)
```

Add a 'count' column with all values set to 1

```python
df['count'] = 1
```

Create age band with age bins using pd.cut

```python
bins = [0, 20, 25, 36, np.inf]
labels = ['Under 20yrs', '20-24yrs', '25-35yrs', 'Above 35yrs']
df['age_band'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

Calculate churn rate and count using group by

```python
churn_stats = df.groupby('churn')['count'].agg(['sum', 'count'])
```

```python
churn_stats['churn_rate'] = round(churn_stats['sum'] / len(df) * 100, 2)
```

Display churn rate, churn counts and age band counts

```python
churn_stats
```

```python
age_band_count = df['age_band'].value_counts()
age_band_count.sort_values(ascending=False)
```

### **Visualization**

```python
# Plot total users gender split
plt.figure(figsize=(4, 4))
gender_counts = df['gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', shadow=False, startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen'],
         wedgeprops=dict(width=0.5, edgecolor='w')) # Adjust slice width and edge color
plt.title("Total Users Gender Split", fontsize=14, fontweight='bold') # Adjust title font size and weight
plt.axis('equal') # Equal aspect ratio ensures a circular pie chart
plt.ylabel('')
plt.tight_layout() # Adjust spacing for better readability
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=25, kde=True, color='skyblue')
plt.title("Distribution of Users by Age", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)  # Prevent overlapping labels
plt.yticks(fontsize=12)

# Remove top and right spines for cleaner design
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add a subtle background color for better contrast
plt.gca().set_facecolor('#f5f5f5')

plt.show()
```

```python
# Calculate churn counts
churn_counts = df['churn'].value_counts()

# Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#66b2ff'])
plt.title("Churned vs Active Users", fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart

plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot churn rate by gender
plt.figure(figsize=(6, 5))
sns.countplot(x='gender', hue='churn', data=df, palette=['skyblue', 'lightcoral'])
plt.title("Churn Rate by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Number of Users", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add faint gridlines for better readability
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title="Churn Status", labels=['Not Churned', 'Churned'], loc='upper left', bbox_to_anchor=(1.05, 1))  # Add legend outside the plot

# Remove top and right spines for cleaner design
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add a subtle background color for better contrast
plt.gca().set_facecolor('#f5f5f5')

plt.show()
```

```python
# Plot Churn Rate by Age Distribution
plt.figure(figsize=(10, 7))
sns.countplot(x='age_band', hue='churn', data=df)
plt.title("Churn Rate by Age Distribution")
plt.show()
```

```python
# Calculate churn rate for each year
df_grouped = df.groupby(df['joined_date'].dt.year)['churn_rate'].mean() * 100

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x=df_grouped.index, y=df_grouped.values)

# Customize the plot for clarity and aesthetics
ax.set_title("Churn Rate by Year Joined", fontsize=14, fontweight='bold')
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Churn Rate (%)", fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)  # Add faint gridlines for readability
ax.xaxis.set_major_locator(plt.MultipleLocator(1))  # Ensure spacing between years

# Optional emphasis on churned bars
for p in ax.patches:
    if p.get_label() == "Churned":
        p.set_alpha(0.8)  # Make churned bars slightly transparent
        p.set_edgecolor('black')  # Add black border to churned bars
        p.set_linewidth(1)  # Increase border width for churned bars

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plot churn rate by joined month
plt.figure(figsize=(10, 6))
sns.countplot(x=df['joined_date'].dt.month_name(), hue='churn', data=df, palette=['#fab400', '#008080'])

# Improve title clarity and relevance
plt.title("Distribution of Churn Across Join Months", fontsize=16)

# Customize labels and grid
plt.xlabel('Month Joined', fontsize=14)
plt.ylabel('Number of Customers', fontsize=14)
plt.tick_params(labelsize=12)

# Enhance legend
plt.legend(title="Churned", labels=["No", "Yes"], title_fontsize=14, loc='upper left', bbox_to_anchor=(1.05, 1))

# Add subtle grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Remove top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# Calculate churn rate by gender
churn_rate = (df.groupby('gender')['churn']
              .value_counts(normalize=True) * 100.0)

# Plot churn rate as a bar chart
fig, ax = plt.subplots(figsize=(10, 6))
churn_rate.unstack('churn').plot.bar(stacked=True, color=['#fab400', '#008080'], ax=ax)

# Customize title, labels, and grid
font1 = {'color': 'black', 'size': 16}
plt.title("Churn Rate by Gender", fontdict=font1)
ax.set_ylabel('Churn Rate (%)', size=14)
ax.set_xlabel('Gender', size=14)
plt.tick_params(labelsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Enhance legend
plt.legend(title="Churned", labels=["No", "Yes"], title_fontsize=14, loc='upper left', bbox_to_anchor=(1.05, 1))

# Add data annotations above each bar
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    ax.annotate('{:.1f}%'.format(height), (p.get_x() + 0.2 * width, p.get_y() + 0.4 * height),
                fontsize=12, color='white', weight='bold')

# Remove top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Additional customizations for a cleaner look
plt.tight_layout()  # Adjust spacing between elements for a more balanced look

plt.show()
```

Females have a higher churn rate to males

```python
# Calculate churn rate by year and month
heatmap_data = df.groupby([df['joined_date'].dt.year, df['joined_date'].dt.month])['churn_rate'].mean().unstack()

# Create the heatmap
plt.figure(figsize=(12, 4))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2%", cbar_kws={'label': 'Churn Rate'})

# Customize the plot
plt.title("Churn Rate Heatmap by Year and Month", fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Year', fontsize=14)
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=12)  # Adjust y-axis label size

# Additional customizations for a cleaner look
plt.grid(False)  # Remove grid lines
plt.tick_params(bottom=False, left=False)  # Remove ticks on bottom and left axes
plt.gca().set_aspect('equal')  # Ensure a square heatmap

# Show the plot
plt.tight_layout()
plt.show()
```
